Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def transformation_of_experience(x):
    if (x is np.nan) or ("Не указано" in x):
        return np.nan
    str_1 = x.split()[:6]
    year_1 = 0
    month_1 = 0
    for index, item in enumerate(str_1):
        if ('год' in item) or ('лет' in item):
            year_1 = int(str_1[index-1])
        elif 'месяц' in item:
            month_1 = int(str_1[index-1]) 
    return int(year_1 * 12 + month_1)

database_hh['Опыт работы (месяц)'] = database_hh['Опыт работы'].apply(transformation_of_experience)
print('Медианный опыт работы ',database_hh['Опыт работы (месяц)'].median())
database_hh = database_hh.drop('Опыт работы', axis=1)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    * … , готов к переезду , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def city_1(x):
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    x = x.split(',')
    x = x[0].strip()
    if x == 'Москва':
        return 'Москва'
    elif x == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif x in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

def relocation_1(x):
    x = x.split(',')
    for index, item in enumerate(x):
        if ('перее' in item):
            if (('готов' in item) or ('хочу' in item)) and ('не' not in item):
                return True
    return False

def business_trip_1(x):
    x = x.split(',')
    for index, item in enumerate(x):
        if ('командиров' in item):
            if ('готов' in item) and ('не' not in item):
                return True
    return False

database_hh['Город'] = database_hh['Город, переезд, командировки'].apply(city_1)
database_hh['Готовность к переезду'] = database_hh['Город, переезд, командировки'].apply(relocation_1)
database_hh['Готовность к командировкам'] = database_hh['Город, переезд, командировки'].apply(business_trip_1)

print('Процент соискателей в Санкт-Петербург:',round(database_hh['Город'].value_counts(normalize=True)['Санкт-Петербург']* 100)) 
print('Процент соискателей готовых к перезду:', round(database_hh[database_hh['Готовность к переезду'] & database_hh['Готовность к командировкам']].shape[0] / database_hh.shape[0] *100))

database_hh = database_hh.drop('Город, переезд, командировки', axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
employment_1 =['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
work_schedule_1 =['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

for index, item in enumerate(employment_1):
    database_hh[item] = database_hh['Занятость'].apply(lambda x: item in x)

for index, item in enumerate(work_schedule_1):
    database_hh[item] = database_hh['График'].apply(lambda x: item in x)

print('Людей ищут проектную работу и волонтерство:', database_hh[database_hh['проектная работа'] & database_hh['волонтерство']].shape[0])
print('Людей хотят работать вахтовым методом и с гибким графиком: ', database_hh[database_hh['вахтовый метод'] & database_hh['гибкий график']].shape[0])

database_hh = database_hh.drop(['Занятость','График'], axis=1)
```

```python
exchange_rates = pd.read_csv('data/ExchangeRates.csv', sep=",")
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'], format='%d/%m/%y').dt.date
database_hh['Обновление резюме'] = pd.to_datetime(database_hh['Обновление резюме']).dt.date

def salary_conversion(x):
    x = x.split()
    x = x[0].strip()
    return int(x)

def currency_conversion(x):
    dictionary_of_currency = {  'RUB': 'руб.',
                                'KZT': 'KZT',
                                'USD': 'USD', 
                                'BYN': 'бел.руб.',
                                'EUR': 'EUR',
                                'UAH': 'грн.',
                                'UZS': 'сум',
                                'KGS': 'KGS',
                                'AZN': 'AZN'}
    
    x = x.split()
    x = x[1].strip() 
    for key in dictionary_of_currency:
        if x == dictionary_of_currency[key]:
            return key
    

database_hh['Желаемый заработок'] = database_hh['ЗП'].apply(salary_conversion)
database_hh['Валюта'] = database_hh['ЗП'].apply(currency_conversion)

tmp_data_frame = database_hh.merge(
    exchange_rates, 
    left_on=['Валюта', 'Обновление резюме'],
    right_on=['currency','date'], 
    how='left')

tmp_data_frame['close'] = tmp_data_frame['close'].fillna(1)
tmp_data_frame['proportion'] = tmp_data_frame['proportion'].fillna(1)

database_hh['ЗП (руб)'] = database_hh['Желаемый заработок'] * tmp_data_frame['close'] / tmp_data_frame['proportion']

database_hh = database_hh.drop(['Желаемый заработок','Валюта', 'ЗП'], axis=1)

display(database_hh)
display(database_hh.info())

print('Медианная желаемая зарплата соискателей: ', database_hh['ЗП (руб)'].median())
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

database_hh_clean_tmp1 = database_hh_clean.copy()

fig, ax = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(database_hh_clean_tmp1['Возраст'] + 1)
histplot = sns.histplot(log_age, bins=20, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 4 * log_age.std(), color='red', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='green', ls='--', lw=2)
histplot.set_title('Распределение признака возраст распределение признака в логарифмическом масштабе');


outliers, database_hh_clean = outliers_z_score_mod(database_hh_clean_tmp1, 'Возраст', log_scale=True, left=3, right=4)

print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {database_hh_clean.shape[0]}')
display (outliers)
```

По методу распределения 3 сигм надено 3 выброса (резюме соискателй с возрастом 15,15 и 100 лет).
